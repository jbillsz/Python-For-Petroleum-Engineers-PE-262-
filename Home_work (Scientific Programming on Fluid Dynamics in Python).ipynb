```python
import numpy as np # importing numpy under the alias np
from matplotlib import pyplot as plt #importing pyplot from matplotlib under the alias plt
#Create variables and assign data
rho = 1000 #Density of water is 1000 kg/m3
D = 16.3 #Diameter of pipe is 16.3m
mu = 0.001 #viscosity of water is 1cp = 0.001 kg/ms
#Generate an array containing the values of average flow velocity
v_avg = np.arange(0.000001,0.0002,0.000001)
#Compute the Reynolds Number
Re = (rho*v_avg*D)/mu
#Writing a for loop to loop from Re array and append to the empty lists laminar and turbulent
laminar=[]
turbulent=[]
for i in Re:
    if i <=2100:
        lam=0.05*i*D
    elif i>2100:
        turb=1.359*(i**0.25)*D
        turb
    else:
        print(None)
    laminar.append(lam)
    turbulent.append(1.359*(i**0.25)*D)
len(laminar),len(v_avg),len(turbulent), len #checking if the dimensions of the lists match is the same as the average velocity
```

```python
plt.plot(v_avg,laminar) #plots laminar hydrodynamic length on the y axis against average velocity
plt.xlabel('Average velocity')# this method or function labels the x-axis as average velocity
plt.ylabel('Laminar') # this method or function labels the y axis as Laminar to represent the val
plt.xticks(rotation=45) #Rotate the tick markers by 45 degrees so the values on the x axis can be seen properly
plt.show()# displays the output of the graph.
```

```python
plt.plot(v_avg,turbulent)
plt.xticks(rotation=90)#Rotate the tick markers by 90 degrees so the values on the x axis can be seen properly
plt.xlabel('Average Velocity')
plt.ylabel('Turbulent hydrodynamic length')
#plt.show()
```

```python
plt.plot(v_avg,laminar, label='Laminar Hydro dynamic length') #same as the plot function but the label argument helps when creating a legend for multiple plots on the same graph
plt.plot(v_avg,turbulent,label='Turbulent Hydro dynamic length')
plt.legend() #creates a legend
plt.xticks(rotation=90)
plt.xlabel('Average velocity')
plt.ylabel('Lamnar and Turbulent hydrodynamic Length')
plt.show()
```
